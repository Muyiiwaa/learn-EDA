## COMMON PANDAS METHODS

1. head -- first n rows
2. tail -- last n rows
3. shape -- get no of rows and columns
4. astype -- converts datatypes
5. describe -- for getting summary statistics
6. drop -- dropping rows or columns
7. group by -- for grouping
8. isnull.sum -- for checking null values
9. dropna -- for dropping null values
10. apply -- for fuction calling
11. reset_index -- for reseting index
12. value_counts -- for counting unique values and their frequency

```python
data = data[data['Income'] != ' ']
```

```python
# .astype method

data = data.astype(
    {'Income':'int64',
     'Dt_Customer':'datetime64[ns]'}
)
```

```python
# describe method

data[['Income','Recency']].describe()
```

```python
data.drop([0,1,2], inplace = True)
```

```python
# rest_index

data.reset_index(inplace = True, drop = True)
```

```python
# dropping columns

data.drop(['ID','Year_Birth'], axis = 1, inplace = True)
```

```python
data[['Income','Education']].groupby(['Education']).mean()
```

```python
data['Marital_Status'].value_counts()
```

```python
# row editing with list comprehension

data['Marital_Status'] = ['Single' if x in ['Alone','Absurd','YOLO','Single'] else x for x in data['Marital_Status']]
```

```python
data.isnull().sum()
```

```python
# create the tax function
def get_tax(x):
  tax = x * 0.04
  return tax

# call the function using .apply
data['income_tax'] = data['Income'].apply(get_tax)
```
