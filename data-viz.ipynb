### Data Visualization

```python
# importing the libraries

# data handling libraries
import pandas as pd
import numpy as np

# libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# customizing the visuals
sns.set_style('darkgrid')
sns.set_palette('husl')
```

```python
# load the dataset

orders_df = pd.read_csv('orders.csv')
```

```python
# generate a revenue column

orders_df['revenue'] = orders_df['unitprice'] * orders_df['quantity']
```

### Bar Chart



```python
# seaborn

sns.barplot(data = orders_df, y = 'staff', x = 'revenue',estimator= 'sum')

plt.title('Total revenue by each staff')
plt.xlabel('Total Revenue')
plt.ylabel('Name of Staff')
```

```python
# plotly

px.histogram(data_frame= orders_df, x = 'staff', y = 'revenue',
       title= 'Bar Chart of Revenue by staff', 
       histfunc= 'avg', color= 'shipper', barmode= 'group')
```

### ScatterPlot

```python
# seaborn

sns.scatterplot(data= orders_df, x = 'quantity', y = 'revenue',
                hue= 'shipper')
```

```python
# plotly

px.scatter(data_frame= orders_df, x = 'quantity', y ='revenue',
           color = 'shipper')
```

### histogram, boxplot and other Distribution plot

```python
# seaborn -- single
sns.boxplot(data = orders_df, x = 'unitprice')
```

```python
# seaborn -- multiple


sns.boxplot(data = orders_df, y = 'staff', x = 'revenue')
```

```python
# plotly

px.box(data_frame=orders_df, x = 'shipper', y = 'unitprice',
       color= 'shipper')
```

```python
# histogram -- seaborn

sns.histplot(data= orders_df, x = 'freight', kde = True )
```

```python
px.histogram(data_frame=orders_df, x = 'freight', marginal='box')
```

```python
# line chart

sns.lineplot(data = orders_df,  x = 'orderdate', y = 'quantity')
```

```python
px.line(data_frame=orders_df, x = 'orderdate', y= 'quantity')
```
